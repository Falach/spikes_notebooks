```python
# lowpass filter 70 UCLA
from mnelab.io.writers import write_edf
ucla_subjects = ['396', '398', '400', '402', '404', '405', '406', '414', '415', '416', '417', '422', '423', '426', '429']
conf = {'396': ['LAH1'], '422': ['RAH1'], '426': ['RA1', 'LA1']}
lower_sr = {'394': 124, '422': 99}
for subj in [ '426']:
    depth =  ['LAH1', 'RAH1']
    depth_h = 499
    if subj in conf.keys():
        depth = conf[subj]
    if subj in lower_sr.keys():
        depth_h = lower_sr[subj]
    scalp = ['EOG1', 'EOG2']
    raw = mne.io.read_raw_edf(f'C:\\UCLA\\P{subj}_overnightData.edf')
    raw.pick_channels(depth + scalp)
    raw.load_data()
    # zero phase prevent delay
    raw.filter(l_freq=0.1, h_freq=depth_h, picks=depth, phase='zero-double')
    raw.notch_filter((60, 120, 180, 240), method='spectrum_fit', phase='zero-double')
    raw.filter(l_freq=0.1, h_freq=70, picks=scalp, phase='zero-double')
    write_edf(f'C:\\UCLA\\P{subj}_filtered_70.edf', raw)
```
